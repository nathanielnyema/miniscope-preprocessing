```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
corr = np.load('adjacency.npy')
plt.matshow(corr);
```

```python
G=nx.from_numpy_matrix(corr)
deg = pd.Series(dict(G.degree))
deg = deg.loc[deg>0]
nx.draw(G,  nodelist = deg.index.tolist(), 
        edge_cmap= plt.cm.Reds, cmap= plt.cm.Reds,
        edge_color = nx.get_edge_attributes(G, 'weight').values(), 
        node_size = 50, node_color = np.log(deg).values)
```
